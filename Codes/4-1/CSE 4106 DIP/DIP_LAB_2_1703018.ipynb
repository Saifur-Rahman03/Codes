```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
```

# Input Processing

White Background Image

```python
img1_wh = np.zeros((500, 500, 3), dtype= "uint8")
img1_wh.fill(255)
cv2_imshow(img1_wh)
print('\n')
img2_wh = np.zeros((500, 500, 3), dtype= "uint8")
img2_wh.fill(255)
cv2_imshow(img2_wh)
```

Add a circle

```python
img1 = cv2.circle(img1_wh, (250, 250), 250, (0, 0, 0), -1)
cv2_imshow(img1)
```

Add a rectangle

```python
img2 = cv2.rectangle(img2_wh, (0, 0), (500, 500), (0, 0, 255), -1) # -1 to fill image
cv2_imshow(img2)
```

# Arithmetic Operations

Addition

```python
dst = cv2.addWeighted(img1, 1, img2, 1, 0)
cv2_imshow(dst)
```

```python
img_add = cv2.add(img1, img2)
cv2_imshow(img_add)
```

Subtraction

```python
img_sub = cv2.subtract(img1, img2)
cv2_imshow(img_sub)
```

Multiplication

```python
img_mul = cv2.multiply(img1, img2)
cv2_imshow(img_mul)
```

# Logical Operations

AND

```python
img_and = cv2.bitwise_and(img1, img2)
cv2_imshow(img_and)
```

OR

```python
img_or = cv2.bitwise_or(img1, img2)
cv2_imshow(img_or)
```

NOT

```python
img_not = cv2.bitwise_not(img1, img2)
cv2_imshow(img_not)
```

# Grayscale Operations

Read Image

```python
img = cv2.imread('/content/human_image.jpg') #read image
cv2_imshow(img) # display image
```

GrayScale Image

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) # convert to grayscale
cv2_imshow(gray)
```

Mean Intensity

```python
height = gray.shape[0]
width = gray.shape[1]
c = 0
sum = 0

for i in range(0, height):
  for j in range(0, width):
    sum += gray[i,j]
    c = c + 1
    
avg = sum / c
print('Mean Intensity: ', avg)
```

Plot Image After converting to grayscale

```python
img_watch = cv2.imread('/content/watch_image.jpg') #read image
img_copy = img_watch.copy()
height = img_copy.shape[0]
width = img_copy.shape[1]
c = 0
sum = 0

for i in range(0, height):
  for j in range(0, width):
    x = 255 - img_copy[i, j]
    img_copy[i, j] = x
    
plt.subplot( 1, 2, 1 ), plt.imshow( img_watch)
plt.subplot( 1, 2, 2 ), plt.imshow( img_copy)

plt.show()
```
